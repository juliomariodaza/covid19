```python
import pandas as pd
import plotly.graph_objects as go
def Graf_vs(Paises):
    datos=pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
    datos.drop([datos.columns[0],datos.columns[2],datos.columns[3]], axis=1, inplace=True)
    datos=datos.groupby(datos.columns[0]).sum()
    datos.index=datos.index.str.lower()
    pais=[]
    fig=go.Figure()
    for i in range(len(Paises)):
        pais.append(datos[datos.index==Paises[i]].T)        
        fig.add_trace(go.Scatter(x=pais[i].index,y=pais[i][pais[i].columns[0]],line_shape='linear',name=Paises[i])) 
    return fig.show()

paises=["argentina","chile","peru","colombia","ecuador"] #Escribe los paises que quieras graficar
Graf_vs(paises)
```
